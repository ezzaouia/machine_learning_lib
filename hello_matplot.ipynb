```python
# A bit of import
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style

%matplotlib inline

style.use('ggplot')
```

```python
X = [1, 3, 5]
Y = [3, 5, 6]
X2 = [0.5, 2, 5]
Y2 = [3, 3, 6]

plt.plot(X, Y, '-o', label='line one', linewidth=3)
plt.plot(X2, Y2, '-o' ,label='line two', linewidth=3)

plt.legend(['train', 'val'], loc='upper left')
#plt.legend()

plt.grid(True,color='k')

plt.title("hello matplot")
plt.xlabel("X title")
plt.ylabel("X title")

plt.show()
```

```python
x = [5,8,10]
y = [12,16,6]

x2 = [6,9,11]
y2 = [6,15,7]

plt.scatter(x, y, label='one' , color='m')#, align='center')

plt.scatter(x2, y2, label='two', color='g')#, align='center')

plt.legend()

plt.title('Epic Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

```python
x = [5,8,10]
y = [12,16,6]

x2 = [6,9,11]
y2 = [6,15,7]


plt.bar(x, y, align='center')

plt.bar(x2, y2, color='g', align='center')

plt.legend(['one', 'two'], loc='upper right')

plt.title('Epic Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

```python
def sigmoid(x):
    return 1.0 / (1 + np.exp(x))

x = np.linspace(-5,5,100)

plt.plot(x, sigmoid(x), marker='v', label='Sigmoid')
plt.legend()

plt.ylabel("Sinus")
plt.xlabel("Axis X")

plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
plt.figure(1)
plt.subplot(3,1,1) # nbre de lignes, nbre de colonnes, numéro de la figure
                   # nombre de lignes multiplié par le nombre de 
                   # colonnes est supérieur ou égal au nombre de figure
plt.plot(t1, f(t1), 'bo', t2, f(t2), color='b')
plt.axis('off')
plt.subplot(3,1,2)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.axis('off')
plt.subplot(3,1,3)
plt.plot(t2, np.sin(2*np.pi*t2), 'b-')
plt.axis('off')
```
